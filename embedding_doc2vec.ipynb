```python
import pandas as pd
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import collections
from gensim.models import doc2vec
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('C:/Users/dowoon/Desktop/리영 유사도/all_after_filter.csv')
```

```python
person_id=data['PERSON_ID'].unique()
```

```python
# 날짜 별로 정리
data.sort_values('RECU_FR_DT',inplace=True)
```

```python
# 무의미한 columns 제거
data.drop(columns=['KEY_SEQ','RECU_FR_DT','MAIN_ICDnorm'],inplace=True)
data.drop(columns=['CLAUSE_CD','ITEM_CD','SEQ_NO','SEQ_NO_dise','SEQ_NO_proce','SICK_SYM'],inplace=True)
```

```python
# 환자 id에 따라 문서 구성
all_doc=[]
features=['SEX','AGE_GROUP','CLA_ITE','GNL_NM_CD','SICK_SYMnorm']
for tag,index in enumerate(person_id):
    for i in data[data.PERSON_ID==index][features].iterrows():
        thisdoc=i[1].tolist()
        thisdoc=','.join('%s' %con for con in thisdoc)
        thisdoc=thisdoc.split(',')
        all_doc.append(TaggedDocument(thisdoc, [tag]))
len(all_doc)
all_doc[0]
```

```python
all_doc
```

```python
X_train,X_test=train_test_split(all_doc,test_size=0.2)
```

```python
#doc2vec
model = Doc2Vec(vector_size=128, dm=1,window=2, min_count=2,workers=30,epochs=80)
```

```python
model.build_vocab(X_train)
```

```python
%time model.train(X_train, total_examples=len(X_train), epochs=model.epochs)
```

```python
# 모델 저장
model.save('C:/Users/dowoon/Desktop/리영 유사도/p2v_128_all_1.model')
```
